How many teams won more than 20 games this past season?


```python
wins20 = df[df['W'] > 20]
wins20.TEAM
```

Draw a histogram of the W column.

```python
fig = px.histogram(df, x = 'W', nbins = 25, title = "Distribution of Wins")

fig.show()
```

In the Big 10 Conference, how many games were won by each team? 

```python
fig = px.bar(
    df[df['CONF']=="B10"],
    x = 'TEAM',
    y = "W",
    title = "Number of Ws By Team in the Big 10"
)

fig.show()
```

What percentage of the overall pool of men's basketball teams does each conference in the NCAA represent (as a percentage)?

```python
conf = df[['CONF', 'TEAM']]

fig = px.pie(
    conf,
    names = "CONF",
    title = "Top 10 Teams Win Breakdown",
    width = 800,
    height = 800  
)

fig.show()
```

For teams that won a certain amount of games, what were their average 3-point percentages on offense and defense?

```python
df.groupby("W", as_index = False, dropna = True).agg({"3P_O": 'mean', "3P_D": "mean"})
```

How many games, on average, were played and won by each conference?

```python
df.groupby("CONF", as_index = False, dropna = True).agg({"G":'mean', "W":"mean"})
```

Sort the data by number of effective field goal percentage on offense and defense.

```python
df.sort_values(['EFG_O', 'EFG_D'])
```

Create a 3D Scatter Plot which shows SEC teams and their ORBs and DRBs.

```python
sec = df[df['CONF'] == 'SEC']

fig = px.scatter_3d(
    sec,
    title='SEC Teams, Offensive Rebounds, and Defensive Rebounds 3D Scatter Plot',
    x="TEAM",
    y="ORB",
    z="DRB",
    color="TEAM",
    template='plotly_dark',
    width=800,
    height=800
)
fig.show()
```
